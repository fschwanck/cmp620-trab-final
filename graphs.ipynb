```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
GRAPHS_PATH = "./graphs"
RESULTS_PATH = "./results"
#RESULTS_CPU_FILE = os.path.join(RESULTS_PATH, 'results-cpu.csv')
#RESULTS_CUDA_FILE = os.path.join(RESULTS_PATH, 'results-cuda.csv')
RESULTS_FILE = os.path.join(RESULTS_PATH, 'results.csv')
MODELS_FILES = os.path.join(RESULTS_PATH, 'models','*.csv')

```

```python
os.makedirs(GRAPHS_PATH, exist_ok=True)
```

```python
results_df = pd.read_csv(RESULTS_FILE, index_col=False)
results_df['inference'] = results_df['inference'] * 1000
results_df
```

CPU Graphs

```python

cpu_results = results_df[results_df['device'] == 'cpu']
# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(cpu_results['inference'], cpu_results['accuracy'], alpha=0.7, edgecolors='w', s=100)
# Create a scatter plot with different colors for each category
categories = cpu_results['name'].unique()
colors = plt.cm.get_cmap('tab10', len(categories))

for i, category in enumerate(categories):
    subset = cpu_results[cpu_results['name'] == category]
    plt.scatter(subset['inference'], subset['accuracy'], 
                label=category, 
                color=colors(i), 
                alpha=0.7, 
                edgecolors='w', 
                s=100)

# Adding labels and title
plt.xlabel('Inference Time (ms)')
plt.ylabel('Accuracy')
plt.title('Scatter Plot of Inference Time vs. Accuracy by Model using CPU')

# Adding a legend to identify different categories
plt.legend(title='Model')

# Show plot
plt.grid(True)
plt.savefig(os.path.join(GRAPHS_PATH, 'inference-accuracy-cpu.png'), format='png', dpi=300) 
plt.show()
```

```python
# Use glob to find all CSV files in the directory
csv_files = glob.glob(MODELS_FILES)

# List to hold individual DataFrames
df_list = []

# Loop through the list of CSV files and read each into a DataFrame
for file in csv_files:
    df = pd.read_csv(file)
    df_list.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(df_list, ignore_index=True)
combined_df['inference_time'] = combined_df['inference_time'] * 1000

# Display the combined DataFrame
combined_df
```

```python
cpu_df = combined_df[combined_df['device'] == 'cpu']

# Create a line plot
plt.figure(figsize=(12, 8))

# Plot each model's accuracy as a line
for model in cpu_df['model'].unique():
    subset = cpu_df[cpu_df['model'] == model]
    plt.plot(subset['index'], subset['accuracy'], 
             label=model, 
             marker='s', 
             linestyle='-', 
             linewidth=0.2, 
             markersize=2)

# Adding labels and title
plt.xlabel('Image Count')
plt.ylabel('Accuracy')
plt.title('Line Graph of Accuracy by Model (CPU)')

# Adding a legend
plt.legend(title='Model')

# Show grid
plt.grid(True)

# Save the plot to a file
#output_file_path = 'path/to/save/line_plot.png'  # Change the path and filename as needed
plt.savefig(os.path.join(GRAPHS_PATH, 'models-accuracy-cpu.png'), format='png', dpi=300) 

# Optionally, display the plot
plt.show()
```

```python
gpu_df = combined_df[combined_df['device'] == 'cuda']

# Create a line plot
plt.figure(figsize=(12, 8))

# Plot each model's accuracy as a line
for model in gpu_df['model'].unique():
    subset = gpu_df[gpu_df['model'] == model]
    plt.plot(subset['index'], subset['accuracy'], 
             label=model, 
             marker='s', 
             linestyle='-', 
             linewidth=0.2, 
             markersize=2)

# Adding labels and title
plt.xlabel('Image Count')
plt.ylabel('Accuracy')
plt.title('Line Graph of Accuracy by Model (GPU)')

# Adding a legend
plt.legend(title='Model')

# Show grid
plt.grid(True)

# Save the plot to a file
#output_file_path = 'path/to/save/line_plot.png'  # Change the path and filename as needed
plt.savefig(os.path.join(GRAPHS_PATH, 'models-accuracy-gpu.png'), format='png', dpi=300) 

# Optionally, display the plot
plt.show()
```
